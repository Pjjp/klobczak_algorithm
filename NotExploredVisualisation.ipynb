```python
from IPython.core.display import display, HTML
```

```python
col3_properties = {
    'c_name': 'C',
    'c_type': 'int',
    'areas': [
        {
            'c_min': '28',
            'c_max': '41',
            'c_position': '10%',
            'width': '30%'
        },
        {
            'c_min': '55',
            'c_max': '63',
            'c_position': '20%',
            'width': '20%'
        }
    ]
}
```

```python
def make_column(c_name, c_type, areas):
    
    def make_area(c_min, c_max, c_position, width):
        
        return '''
        <div class="area" style="margin-left: '''+c_position+'''; width: '''+width+''';">
            <div class="bordVal">'''+c_min+'''</div>
            <div class="red"></div>
            <div class="bordVal">'''+c_max+'''</div>
        </div>  
        '''
    
    def make_areas(areas):
        
        result = ''
        for area in areas:
            result = result + make_area(**area)
                
        return result
    
    return '''
    <div class="column">
        <div>'''+c_name+'''</div>
        <div>'''+c_type+'''</div>
        <div class="notExpMain">
            '''+make_areas(areas)+'''
        </div>
    </div>
    '''
```
