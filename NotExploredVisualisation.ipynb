```python
def make_properties(not_explored_chunks, catalogue_item):
    
    columns_properties = []
    col_prop = {}
    
    # chunks columns have to be in same order
    # catalogue item has to have colums in same order as chunks
    for i, col in enumerate(not_explored_chunks[0].borders):
        
        # in catalogue item value min - first, value max - last
        # ci distribution mustn't be short
        ci_col_dist = catalogue_item.spec[i]['distribution']
        ci_col_dist_row_count = len(ci_col_dist)
        
        
        col_prop['c_name'] = col['column']
        col_prop['c_type'] = col['type']
        
        for ch in not_explored_chunks:
            
            ch_min = ch.borders[i]['minimum']
            ch_max = ch.borders[i]['maximum']
            
            for i in range(ci_col_dist_row_count):
                
                if ch_max >= ci_col_dist[i]['value_max']:
                    ch_max_dis_row = i
                    
            for i in reversed(range(ci_col_dist_row_count)):
                
                if ch_min <= ci_col_dist[i]['value_min']:
                    ch_min_dis_row = i
                
            width = ((ch_max_dis_row-ch_min_dis_row+1)/ci_col_dist_row_count)*100
            position = (1/ci_col_dist_row_count)*ch_min_dis_row*100
        
            col_prop['areas'].append({
                        'c_min': ch_min,
                        'c_max': ch_max,
                        'c_position': position,
                        'c_width': width,
                    })
            
        columns_properties.append(col_prop)
    
    return columns_properties
    
```

```python
def make_columns(columns_properties):
    
    def make_column(c_name, c_type, areas):
    
        def make_area(c_min, c_max, c_position, width):

            return '''
            <div class="area" style="margin-left: '''+c_position+'''; width: '''+c_width+''';">
                <div class="bordVal">'''+c_min+'''</div>
                <div class="red"></div>
                <div class="bordVal">'''+c_max+'''</div>
            </div>  
            '''

        def make_areas(areas):

            result = ''
            for area in areas:
                result = result + make_area(**area)

            return result

        return '''
        <div class="column">
            <div>'''+c_name+'''</div>
            <div>'''+c_type+'''</div>
            <div class="notExpMain">
                '''+make_areas(areas)+'''
            </div>
        </div>
        <div class="textcolumn">
        </div>
        '''
    final_result = ''
    for col_prop in columns_properties:
        
        final_result = final_result + make_column(**col_prop)
        
    return final_result
        
```
